## 3. Visualisation 

```python
import geopandas as gpd
import matplotlib.pyplot as plt 
import os 
import pathlib
```

```python
os.chdir('..')
Path = pathlib.Path().resolve()
```

1. Polygon 

```python
small_areas = gpd.read_file(f'{Path}/data/malmo-pop.shp')
small_areas.head()
```

```python
small_areas.crs
```

```python
small_areas.plot()
```

2. Lines

```python
roads = gpd.read_file(f'{Path}/data/malmo-streets.shp')
roads.head()
```

```python
roads.crs
```

```python
roads.plot(color = 'red', linewidth = 0.2, alpha = 0.8)
```

3. Points 

```python
schools = gpd.read_file(f'{Path}/data/preschools.shp')
schools.head()
```

```python
schools.plot(markersize=15, color='black', edgecolor='white', alpha=0.8, marker='P')
```

4. Multi Layer Maps

```python
fig, ax = plt.subplots(1,1, figsize=(12,10))

roads.plot(ax=ax, color = 'red', linewidth = 0.2, alpha = 1)
small_areas.plot(ax=ax, color = 'grey', alpha = 0.8)
schools.plot(ax=ax, markersize=15, color='black', edgecolor='white', alpha=0.8, marker='P')

plt.title('Schools in Malmo', fontsize=40, color ='grey')
ax.axis('off')
plt.axis('equal')
plt.show()
```

5. Choropleth Map 

With GeoPandas files we just need to provide column on which we want to plot. 

```python
small_areas.plot()
```

```python
small_areas.plot(column='Age80_w', figsize=(12,10))
```

We can add additional features to such a maps: 
- legend 
- scheme 
- change color 
- number of class `k` 

```python
fig, ax = plt.subplots(figsize=(12,10)) 

small_areas.plot(column='Age80_w', cmap='Blues', scheme='quantiles', k=6, ax=ax, legend=True)
```

6. Bubble map 

To represents data as bubbles we first need to represent them as a points

```python
small_areas_points = small_areas.copy()
small_areas_points['geometry'] = small_areas_points['geometry'].centroid 
small_areas_points.head()
```

```python
small_areas_points.plot()
```

```python
fig, ax = plt.subplots(figsize=(9,9)) 

small_areas_points.plot(ax=ax, color='blue', markersize='Age80_w')
plt.axis('equal') 

plt.show()
```

We can add visual context to the bubbles: 

```python
fig, ax = plt.subplots(figsize=(9,9)) 

small_areas.plot(ax=ax, color = 'grey', edgecolor='black', linewidth=0.4)
small_areas_points.plot(ax=ax, color='yellow', markersize='Age80_w', alpha=0.5, categorical=False, legend=True) 

ax.axis('off')
plt.axis('equal')

plt.show()
```
